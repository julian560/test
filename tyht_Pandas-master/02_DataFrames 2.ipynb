## Filter with More than One Condition (AND)

```python
ubike_car = pd.read_csv("ubike-utf8.csv")
ubike_car.head(2)
```

```python
a1 = ubike_car['場站區域'] == '大安區'
a2 = ubike_car['平均車輛數'] > 30
```

```python
ubike_car[a1 & a2]
```

```python
a1 = ubike_car['場站區域'] == '大安區'
a2 = ubike_car['平均車輛數'] > 30
a3 = ubike_car['平均空位數'] == 22
```

```python
ubike_car[a1 & a2 | a3]
```

## The `.isin()` Method

```python
ubike_car = pd.read_csv("ubike-utf8.csv")
ubike_car.head(2)
```

```python
a1 = ubike_car['場站區域'] == '大安區'
a1 = ubike_car['場站區域'] == '信義區'
a1 = ubike_car['場站區域'] == '士林區'

ubike_car[a1 | a2 | a3]
```

```python
ubike_car[ubike_car['場站區域'].isin(['大安區','信義區','士林區'])]
```

```python
#另一種寫法
mask = ubike_car['場站區域'].isin(['大安區','信義區','士林區'])
```

```python
ubike_car[mask]
```

## The `.duplicated()` Method

```python
#重複
ubike_car = pd.read_csv("ubike-utf8.csv")
ubike_car.head()
```

```python
ubike_car['場站區域'].duplicated(keep = 'first')
```

```python
mask = -ubike_car['場站區域'].duplicated(keep = 'first')
#keep = 'first'：除了第一次出現外，其餘相同的被標記為重複
#keep = 'last'：除了最後一次出現外，其餘相同的被標記為重複
#keep = False：所有相同的都被標記為重複
```

```python
ubike_car[mask]
```

```python
ubike_car.drop_duplicates(subset=['場站區域'], keep='first')
```
