## The `.get_level_values()` Method

```python
ubike_car = pd.read_csv("ubike-utf8.csv", parse_dates=['日期'], index_col =  ['日期','場站區域'])
```

```python
ubike_car.sort_index(inplace = True)
ubike_car.head(3)
```

```python
ubike_car.index.get_level_values(0)
```

```python
ubike_car.index.get_level_values("日期")
```

```python
ubike_car.index.get_level_values("場站區域")
```

## The `.set_names()` Method on MultiIndex

```python
ubike_car = pd.read_csv("ubike-utf8.csv", parse_dates=['日期'], index_col =  ['日期','場站區域'])
ubike_car.sort_index(inplace = True)
ubike_car.head(3)
```

```python
ubike_car.index.set_names(["日期", "場站區域內容"], inplace = True)
```

```python
ubike_car.head(3)
```

## Extract Rows from a `MultiIndex DataFrame`

```python
ubike_car.loc[("2014-12-08")]
```

```python
ubike_car.loc[("2014-12-08", "蘆洲區"), '場站名稱']
```

```python
ubike_car.ix[("2014-12-08", "蘆洲區"), 2]
```

## The `pivot` Method

```python
ubike_car = pd.read_csv("ubike-utf8.csv", parse_dates=['日期'])
ubike_car["場站區域"] = ubike_car["場站區域"].astype("category")
ubike_car.head(3)
```

## The `.stack()` Method

```python
ubike_car.stack()
```

```python
ubike_car.stack().to_frame()
```
