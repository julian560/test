## Broadcasting Operations

```python
raw_ubike_car = pd.read_csv("ubike-utf8.csv")
```

```python
raw_ubike_car['平均車輛數'].add(5).head()
```

```python
#Plus加
raw_ubike_car['平均車輛數'] + 10
```

```python
#減(Subtraction)
raw_ubike_car['平均車輛數'].sub(100)
raw_ubike_car['平均車輛數'] - 100
```

```python
#乘(Multiplication)
raw_ubike_car['平均車輛數'].mul(2)
```

```python
raw_ubike_car['平均車輛數'] = raw_ubike_car['平均車輛數'].mul(2)
```

```python
raw_ubike_car['平均車輛數'].head()
```

```python
#除(Division)
raw_ubike_car['平均車輛數'].div(10).head()
```

```python
raw_ubike_car.sort_values(['平均車輛數','最大車輛數'], ascending=False)
```

## Rank Values with the `.rank()` Method

```python
raw_ubike_car['平均空位數'].rank(ascending = False).astype('int')
```

```python
raw_ubike_car['平均空位數排名'] = raw_ubike_car['平均空位數'].rank(ascending = False).astype('int')
```

```python
raw_ubike_car.head()
```

```python
raw_ubike_car.sort_values(by = '平均空位數排名')
```
