## A Review of the `.apply()` Method on Single Columns

```python
ubike_car = pd.read_csv("ubike-utf8.csv", index_col='場站名稱')
ubike_car.sort_index(inplace=True)
ubike_car.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    number = round(number, 2)
    return str(number) + '輛'
```

```python
columns = ["平均車輛數", "最大車輛數", "最小車輛數"]
for col in columns:
    ubike_car[col] = ubike_car[col].apply(convert_to_string_and_add_millions)
```

```python
ubike_car.head(3)
```
