## Methods on the Groupby Object and `DataFrame` Columns

```python
ubike_car = pd.read_csv("ubike-utf8.csv", index_col='日期')
sectors = ubike_car.groupby("場站區域")
ubike_car.head()
```

```python
sectors.get_group("大安區")["最大空位數"].sum()
```

```python
#Make a maximum value
sectors.max()
#Make a minimum value
sectors.min()
sectors.sum()
sectors.mean()
```

```python
sectors["平均車輛數"].sum()
sectors["最大空位數"].max()
sectors["最大空位數"].min()
sectors["最大空位數"].mean()

sectors[["平均車輛數", "最大空位數"]].sum()
```

## Grouping by Multiple Columns

```python
ubike_car = pd.read_csv("ubike-utf8.csv", index_col='日期')
sectors = ubike_car.groupby(["場站區域",'場站名稱'])
ubike_car.head()
```

```python
sectors.size()
sectors.sum()
sectors["平均空位數"].sum()
sectors["平均車輛數"].mean()
```

## The `.agg()` Method

```python
ubike_car = pd.read_csv("ubike-utf8.csv", index_col='日期')
sectors = ubike_car.groupby(["場站區域",'場站名稱'])
ubike_car.head()
```

```python
sectors["平均車輛數"].mean()
```

```python
sectors.agg({"平均車輛數" : ["sum", "mean"],
             "平均空位數" : "sum",
              "總停車格" : "mean"})
```
